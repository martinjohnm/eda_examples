```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
df.columns
```

```python
plt.scatter(x=df[['SepalLengthCm']],y=df[['SepalWidthCm']])
plt.show()
```

```python
# pair plot 
features=[ 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']
sns.set_style('whitegrid')
sns.pairplot(df[features],hue='Species',height=3)
plt.show()
```

## Observations 
1) Petal length and petal width are the most important feature to distinguish between various flower types 

2) Setosa can be easily clasdified whereas virginica and versicolor have some overlap

```python
df.describe()
```

```python
g=df.groupby('Species')

df_setosa=g.get_group('Iris-setosa')
df_setosa.describe()
```

```python
df_virginica=g.get_group('Iris-virginica')
df_virginica.describe()
```

```python
df_versicolor=g.get_group('Iris-versicolor')
df_versicolor.describe()
```

```python
# Box plots for outlier detection

sns.boxplot(x='Species',y='PetalLengthCm',data=df)
plt.show()
```

```python
sns.violinplot(x='Species',y='PetalLengthCm',data=df)
plt.show()
```

```python
#Countour plotss

sns.jointplot(x='PetalLengthCm',y='PetalWidthCm',data=df_setosa,kind='kde')
plt.show()
```
